```python
import os

# Function to count subfolders with more than 5 images
def count_subfolders_with_more_than_10_images(root_dir):
    count = 0

    # Iterate over subdirectories in the root directory
    for dirpath, dirnames, filenames in os.walk(root_dir):
        # Count the number of image files in the current subdirectory
        num_images = len([f for f in filenames if f.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp'))])

        # Check if the current subdirectory contains more than 5 images
        if num_images >= 10:
            # print(f"Subdirectory '{dirpath}' contains more than 10 images.")
            count += 1

    return count

# Example usage:
root_directory = r"G:\My Drive\Data Mining\Project 1\code test\custom_data"

num_subfolders_with_more_than_10_images = count_subfolders_with_more_than_10_images(root_directory)
print(f"Number of subfolders with more than 10 images: {num_subfolders_with_more_than_10_images}")
```

```python


import os
import shutil

def count_and_copy_subfolders(root_dir, output_dir):
    count = 0
    i = 0 
    # Iterate over subdirectories in the root directory
    for dirpath, dirnames, filenames in os.walk(root_dir):
        # Count the number of image files in the current subdirectory
        num_images = len([f for f in filenames if f.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp'))])

        # Check if the current subdirectory contains more than 10 images
        if num_images >= 10:
            print(f"Subdirectory '{dirpath}' contains more than 10 images.")
            count += 1
            
            # Create a corresponding subfolder in the output directory
            relative_path = os.path.relpath(dirpath, root_dir)
            output_subfolder = os.path.join(output_dir, relative_path)
            os.makedirs(output_subfolder, exist_ok=True)

            # Copy all files from the current subdirectory to the output subfolder
            for filename in filenames:
                src_path = os.path.join(dirpath, filename)
                dest_path = os.path.join(output_subfolder, filename)
                shutil.copyfile(src_path, dest_path)
        # if i == 50: 
        #     break 
    return count

# Example usage:
root_dir = r"G:\My Drive\Data Mining\Project 1\code test\lfw-deepfunneled"
output_dir = r"G:\My Drive\Data Mining\Project 1\code test\custom_data"

num_subfolders = count_and_copy_subfolders(root_dir, output_dir)
print(f"Total {num_subfolders} subfolders with more than 10 images were copied to '{output_dir}'.")



```

```python
import os
import random
import shutil

def delete_random_subfolders(folder_path, n):
    if not os.path.exists(folder_path):
        print(f"Folder '{folder_path}' does not exist.")
        return

    subfolders = [f.path for f in os.scandir(folder_path) if f.is_dir()]
    if len(subfolders) <= n:
        print(f"There are not enough subfolders to delete in '{folder_path}'.")
        return

    subfolders_to_delete = random.sample(subfolders, n)
    for subfolder in subfolders_to_delete:
        shutil.rmtree(subfolder)
        print(f"Deleted subfolder: {subfolder}")

# Example usage:
folder_path = "G:\My Drive\Data Mining\Project 1\code test\custom_data"
number_of_subfolders_to_delete = 10
delete_random_subfolders(folder_path, number_of_subfolders_to_delete)
```
